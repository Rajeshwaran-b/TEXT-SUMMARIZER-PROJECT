```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV output input 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split # splitting the dataset
from sklearn import preprocessing # data processing
from sklearn.preprocessing import Normalizer
import matplotlib.patches as mpatches
from sklearn.metrics import confusion_matrix, accuracy_score

import seaborn as sns

from tensorflow.keras.utils import to_categorical
import tensorflow as tf
from tensorflow.keras import models
from tensorflow.keras.layers import Dense
```

```python
# import data set 
f = open(u'C:/Users/РђРЅРіРµР»РёРЅР°/Downloads/1611955500_bank.csv')
df = pd.read_csv(f, sep=',', encoding='cp1251')
f.close()
```

```python
# drop the columns with unique data for each customer: 
# RowNumber, CustomerId, Surname
df = df.drop('RowNumber', 1)
df = df.drop('CustomerId', 1)
df = df.drop('Surname', 1)
```

```python
df['Gender'].replace({'Male':1,'Female':0}, inplace= True) 
# create dummy for gender
#%%
  
df1= pd.get_dummies(data= df, columns=['Geography']) 
# create dummies for geography

df1.head()

#%%
```

```python
# Bivariate Analysis
# Correlation Matrix

plt.figure(figsize=(12, 8))

df1_corr = df1.corr()
sns.heatmap(df1_corr, 
            xticklabels = df1_corr.columns.values,
            yticklabels = df1_corr.columns.values,
            annot = True);

# It is seen that people of the higher age are more likely to churn. Also
# those people, who are from Germany are more likely to churn as well as those
# who have a higher balance on their account
```

```python
# Bivariate Analysis
# Crossplots

plt.figure(figsize=(14, 14))

sns.pairplot(df1, diag_kind='kde');

# There are no clear patterns that can be seen from the scatter plots.
```

```python
# Creating Train Set & Test set
 
X = df1.drop('Exited',axis='columns') # choosing the regressors - feature set
Y = df1['Exited'] # i have chosen the dependent variable - target set

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=5) 
# split on training and test sets 80%-20%        

X_train.shape

#%%
```

```python
# Normalize the data
normalize=Normalizer(norm="l2")
X_train=normalize.transform(X_train)

print(X_train)

X_test=normalize.transform(X_test)

print(X_test)
#%%
```

```python
# NN

nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=6, activation='relu'))
nn.add(tf.keras.layers.Dense(units=6, activation='relu'))
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
nn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

nn.fit(X_train, Y_train, batch_size = 10, epochs = 100)
```

```python
# Prediction

Y_pred = nn.predict(X_test)
Y_pred = (Y_pred > 0.5)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
#%%

conf_matrix = confusion_matrix(Y_test, Y_pred)
print(conf_matrix)
accuracy_score(Y_test, Y_pred)
```
